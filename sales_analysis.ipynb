```python
import pandas as pd

df= pd.read_csv(r"C:\Users\matey\Desktop\sales analysis\amazon.csv")
```

```python
df.describe(include='all')
```

```python
import numpy as np

price_cols = ['actual_price', 'discounted_price','rating_count']

for col in price_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('₹', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.head()
```

```python
(df['discounted_price'] > df['actual_price']).all()
```

```python
df.columns
```

```python
df['category'].str.count('\|').add(1).value_counts()
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
df['product_type'] = df['category'].str.split('|').str[-1]
```

```python
df.head()
```

```python
df = df.drop(columns=[col for col in df.columns if col.startswith('category_level_')])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['rating_count']=df['rating_count'].fillna(df['rating_count'].median())
```

```python
df.isnull().sum()
```

```python
bins = [0, 20, 50, 100] 
labels = ['Low', 'Medium', 'High']
df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype(float)
df['discount_bucket'] = pd.cut(df['discount_percentage'], bins=bins, labels=labels, include_lowest=True)

print(df[['discount_percentage', 'discount_bucket']].head())
```
